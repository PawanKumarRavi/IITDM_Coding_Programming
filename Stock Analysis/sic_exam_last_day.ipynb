```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime,date,time
```

```python
import pandas_datareader.data as web
import yfinance as yf
import plotly.express as plexp
```

```python
yf.pdr_override()
```

```python
start=datetime.now()-pd.DateOffset(months=3)
end=datetime.now()
```

```python
tickers=['APPL','MSFT','NFLX','GOOG','IBM']
```

```python
dflist=[]
for ticker in tickers:
  data=yf.download(ticker,start=start,end=end)
  dflist.append(data)
```

```python
df=pd.concat(dflist,keys=tickers,names=['ticker','Date'])
df.head(10)
```

```python
df=df.reset_index()
```

```python
img1=plexp.line(
    df,x='Date',y='Close',color='ticker',title='stock price'
)
img1.show()
```

```python
img2=plexp.line(
    df,x='Date',y='Close',color='ticker',title='stock price movements',labels={'Date':'Date','Close':'Close Price','Ticker':'Stock Ticker'}
)
img2.show()
```

```python
df['maverage10']=df.groupby('ticker')['Close'].rolling(window=10).mean().reset_index(0,drop=True)
df['maverage20']=df.groupby('ticker')['Close'].rolling(window=20).mean().reset_index(0,drop=True)


```

```python
df
```

```python
for ticker,group in df.groupby('ticker'):
  print(f"moving average for {ticker}")
  print(group[['maverage10','maverage20']])
  print('\n\n')
```

```python
img3=plexp.line(
    group,x='Date',y=['Close','maverage10','maverage20'],title=f'stock price moving averages for {ticker}'
)
img3.show()
```

```python
df['volatility']=df.groupby('ticker')['Close'].pct_change().rolling(window=10).std().reset_index(0,drop=True)
```

```python
df
```

```python
img4=plexp.line(
    df,x='Date',y='volatility',color='ticker',title='volatility of stocks'
)
img4.show()
```

```python
apple=df.loc[df['ticker']=='AAPL',['Date','Close']].rename(columns={'Close':'AAPL'})
microsoft=df.loc[df['ticker']=='MSFT',['Date','Close']].rename(columns={'Close':'MSFT'})
corr=pd.merge(apple,microsoft,on='Date')

```

```python
img5=plexp.scatter(
    corr,x='AAPL',y='MSFT',title='correlation between apple and microsoft stocks',trendline='expanding'
)
img5.show()
```
